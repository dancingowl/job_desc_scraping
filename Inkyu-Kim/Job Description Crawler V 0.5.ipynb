```python
incruit = pd.DataFrame()
for i in range(1):
    
    # 현재 페이지 소스 파싱
    bs = BeautifulSoup(driver.page_source, 'html.parser')
    posts = bs.find('ul', class_ = 'litype01').find_all('li')
    
    for title in posts:

        Company = title.find('h3').text[:-6] # 기업명
        Title = title.find('p', class_ = 'detail').find('a').text # 제목
        etc = title.find('p', class_ = 'etc').text.split(' | ') # 채용 정보
        EmploymentType = etc[0] # 채용유형
        Area = etc[1] # 근무지역
        Education = etc[2] # 학력
        EmployeeType = etc[3] # 경력여부
        Salary = etc[4] # 연봉
        Deadline = title.find('span', class_ = 'info').text.replace('(', '').replace(')', '').replace('~', '')
        JobPost = title.find('span', class_ = 'rcrtTitle').find('a').get('href')
        if '채용시' in Deadline: pass
        elif '상시' in Deadline: pass
        elif '오늘' in Deadline: Deadline = int(0)
        else:
            Deadline = datetime.strptime(('2020/' + Deadline), '%Y/%m/%d').date()
        CollectDate = datetime.now().date() # D_day
        try:
            D_day = (Deadline - CollectDate).days
        except:
            D_day = np.NaN
        incruit = incruit.append(pd.DataFrame({'Company' : [Company],
                                           'Deadline' : [Deadline],
                                           'Area' : [Area],
                                           'Education' : [Education],
                                           'EmploymentType' : [EmploymentType],
                                           'EmployeeType' : [EmployeeType],
                                           'Title' : [Title],
                                           'JobPost' : [JobPost],
                                           'D_day' : [D_day],
                                           'CollectDate' : [CollectDate],
                                           }))

    driver.find_element_by_xpath('/html/body/div[2]/div[2]/div[1]/div[3]/p[2]/a[5]').click()
    time.sleep(1)

incruit.reset_index(drop = True)
```
